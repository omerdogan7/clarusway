## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">"Task_1 Plotting the Star Wars Characters by gender</p>

<a id="2"></a>
<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
gender_counts = df['gender'].value_counts()
gender_counts
```

```python
# with matplotlib
# Create the plot
bars = plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution', fontsize=14)


# Add count numbers on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}'
             )
plt.show()
```

```python
# with pandas
ax = df['gender'].value_counts().plot.bar()
plt.title('Gender Distribution', fontsize=13,color='r')
# Annotate bars with their counts
for idx, count in enumerate(gender_counts):
    ax.text(idx, count, str(count), ha='center')
plt.show()
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">TASK_3 - Plotting the MASS unde 3 groups: "High_mass", "Normal_mass", "Low_mass" </p>

<a id="4"></a>
<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
# Ensure the column is treated as strings, then remove commas and convert to float
df['mass'] = df['mass'].astype(str).str.replace(',', '').astype(float)

# Then create the categories
df['mass_cat'] = pd.cut(df['mass'], 
                       bins=[-float('inf'), 50, 99, float('inf')],
                       labels=['Low_mass', 'Normal_mass', 'High_mass'])
mass_cat_counts = df['mass_cat'].value_counts(dropna=False)
mass_cat_counts
```

```python
# with pandas
mass_cat_counts = df['mass_cat'].value_counts(dropna=True)
ax = mass_cat_counts.plot.bar()
# Annotate bars with their counts
for idx, count in enumerate(mass_cat_counts):
    ax.text(idx, count, str(count), ha='center')
plt.xticks(rotation=90)    
plt.show()
```

```python
# with matplotlib
bars = plt.bar(mass_cat_counts.index, mass_cat_counts.values)
# Add count numbers on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.0f}', 
             ha='center', va='bottom')

plt.show()
```

#### **Note: Calculate 1 BBY as 1 Year**
1 BBY = 1 Year


```python
df.head()
```

```python
# Step 1: Convert the 'birth_year' column to string
df['birth_year'] = df['birth_year'].astype(str)

# Step 2: Replace 'BBY' with an empty string
df['birth_year'] = df['birth_year'].str.replace('BBY', '').astype(float)

# Step 4: Filter out NaN values and perform the comparison
filtered_df = df[df['birth_year'] >= 100]
filtered_df
```

```python
# with pandas
ax = filtered_df['birth_year'].astype(int).plot.bar(figsize=(8,6), title='Big of Age')
# Annotate bars with their counts
# Add counts at the center of the bars
for p in ax.patches:
    height = p.get_height()
    x = p.get_x() + p.get_width() / 2  # Get the x position of the bar
    y = height / 2  # Position the text at the center of the bar
    ax.text(x, y, str(height), ha='center', va='center',rotation='vertical')
ax.set_xticklabels(filtered_df['name'],rotation=0)
ax.set_ylabel('Age >100')

plt.show()
```

```python
# with matplotlib
# with matplotlib
plt.figure(figsize=(8,6))
bars = plt.bar(filtered_df['name'], filtered_df['birth_year'].astype(int), width=0.4)
# Add count numbers on top of each bar
for bar in bars:
    height = bar.get_height()
    x = bar.get_x() + bar.get_width() / 2  # Get the x position of the bar
    plt.text(x, height,
             f'{height:.0f}', 
             ha='center', va='bottom', rotation='horizontal')
plt.title('Big of Age')
plt.xlabel('Name')
plt.ylabel('Age>100')
plt.xticks(rotation=45)
plt.show()
```
