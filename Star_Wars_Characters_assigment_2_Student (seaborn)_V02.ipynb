**This can now be plotted much more concisely.**<br>
Axes.bar_label() automatically labels bars [since matplotlib 3.4](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar_label.html)

```python
# with seaborn barplot

```

```python
# with seaborn barplot
# Get the value counts for 'gender' and reset the index to create a DataFrame
gender_counts = df['gender'].value_counts().reset_index()

# Create a bar plot using the value counts
sns.barplot(x='gender', y='count', hue='gender', data=gender_counts)
plt.title('Gender', c='b',size=14)
plt.ylabel('')
plt.xticks(rotation=45)
# Adding count number at the top of each bar
for index, row in gender_counts.iterrows():
    plt.text(index, row['count'], str(row['count']), ha='center', va='bottom')
# Display the plot
plt.show()
```

```python
# with seaborn countplot

```

```python
# Create the count plot
ax = sns.countplot(
    x='gender', 
    hue='gender', 
    order=df['gender'].value_counts().index, 
    data=df
)

# Add count numbers on top of each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',  # The count number
                (p.get_x() + p.get_width() / 2, p.get_height()),  # Position at the center-top of the bar
                ha='center', va='center',  # Center alignment
                xytext=(0, 5),  # Offset above the bar
                textcoords='offset points',  # Offset in points
                ) 

# Add a title to the plot
plt.title('Gender', color='blue')

# Customize x-axis tick labels
plt.xticks(rotation=45, color='blue', fontsize=9)
plt.show()
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">TASK_3 - Plotting the MASS under 3 groups: "High_mass", "Normal_mass", "Low_mass" </p>


<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
#mass_cat
```

```python
#mass_cat
# Convert mass to numeric, coerce errors to NaN
df['mass'] = pd.to_numeric(df['mass'], errors='coerce')

# Then create the categories
df['mass_cat'] = pd.cut(df['mass'], 
                       bins=[-float('inf'), 50, 100, float('inf')],
                       labels=['Low_mass', 'Normal_mass', 'High_mass'])
mass_cat_counts = df['mass_cat'].value_counts(dropna=False)
mass_cat_counts
```

```python
# with seaborn barplot

```

```python
# with seaborn barplot
# Create figure with larger size for better readability
plt.figure(figsize=(8, 6))

# Reset the index of counts to prepare for visualization
mass_cat_counts = pd.DataFrame(mass_cat_counts).dropna()

# Create a bar plot
ax = sns.barplot(data=mass_cat_counts.dropna(), 
                 x='mass_cat', y='count', 
                 hue='mass_cat',
                 legend=False,
                 order=['Normal_mass', 'Low_mass', 'High_mass'])

# Customize plot
plt.title('Mass Groups', c='b')
plt.ylabel('')
plt.xticks(rotation=0)

# Add count numbers on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 9),  # 3 points vertical offset
                textcoords='offset points')

plt.show()
```

```python
# With Seaborn Countplot
# Create figure with larger size for better readability and set the background color of the figure
fig, ax = plt.subplots(figsize=(8, 6))

# 'species' is the column of interest
top_species = df['species'].value_counts().nlargest(5).index

# Create the count plot for the top 5 species
sns.countplot(data=df[df['species'].isin(top_species)], x='species', hue='species',order=top_species,ax=ax)

plt.title('Top Species 5')
# Add count numbers inside each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='top',xytext=(0, -6),textcoords='offset points')
plt.show()
```
